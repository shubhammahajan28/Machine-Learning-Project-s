```python
import re
```

```python
def wordopt(text):
    # Convert into lowercase
    text = text.lower()

    #Remove URLS 
    text = re.sub(r'http?://\s+|www\.\S+','',text)

    #HTML
    text= re.sub(r'<.*?','',text)

    # punctuation
    text= re.sub(r'[^\w\s]','',text)

    #Digit
    text = re.sub(r'\d','',text)

    #newline
    text = re.sub(r'\n',' ',text)

    return text
```

```python
news['text'] = news['text'].apply(wordopt)
```
